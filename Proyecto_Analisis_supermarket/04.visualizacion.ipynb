```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo de gráficos
sns.set(style="whitegrid")

# Ruta al archivo CSV limpio
file_path_cleaned = r'C:\Users\Scarlett\Documents\Mi_Proyecto_\Proyecto_Analisis_supermarket\data\supermarket_sales_cleaned.csv'

# Intentar cargar el dataset limpio
try:
    df_cleaned = pd.read_csv(file_path_cleaned)
    print("Dataset limpio cargado correctamente.")
    
    # 1. Distribución de la Calificación (Rating)
    plt.figure(figsize=(10, 6))
    sns.histplot(df_cleaned['Rating'], bins=10, kde=True)
    plt.title('Distribución de la Calificación (Rating)')
    plt.xlabel('Rating')
    plt.ylabel('Frecuencia')
    plt.show()

    # 2. Ventas Totales por Sucursal
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Branch', y='Total', data=df_cleaned)
    plt.title('Ventas Totales por Sucursal')
    plt.xlabel('Sucursal')
    plt.ylabel('Ventas Totales')
    plt.show()

    # 3. Ventas por Línea de Producto
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Total', y='Product line', data=df_cleaned, estimator=sum, ci=None)
    plt.title('Ventas Totales por Línea de Producto')
    plt.xlabel('Ventas Totales')
    plt.ylabel('Línea de Producto')
    plt.show()

    # 4. Ventas por Día de la Semana
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='Day', y='Total', data=df_cleaned, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
    plt.title('Ventas por Día de la Semana')
    plt.xlabel('Día de la Semana')
    plt.ylabel('Ventas Totales')
    plt.show()

    # 5. Relación entre Precio Unitario y Cantidad
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Unit price', y='Quantity', data=df_cleaned)
    plt.title('Relación entre Precio Unitario y Cantidad')
    plt.xlabel('Precio Unitario')
    plt.ylabel('Cantidad')
    plt.show()

except FileNotFoundError:
    print(f"El archivo limpio no se encuentra en la ruta especificada: {file_path_cleaned}")
except Exception as e:
    print(f"Ocurrió un error: {e}")



```
